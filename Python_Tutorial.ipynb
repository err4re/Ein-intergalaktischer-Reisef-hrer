## Bibliotheken importieren

1. `matplotlib` für das plotten/erstellen von Graphen
2. `numpy` für die Numerik/Mathematik

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Lineare Funktion

```python
def lin_function(x, m, t):
    """
    Eine lineare Funktion, welche y = m*x + t berechnet und ausgibt.
    
    Parameter:
        x: Wert aus dem Definitionsbereich
        m: Steigung
        t: y-Achsenabschnitt
        
    Returns:
        y: Funktionswert am Punkt x
        
    """
    
    return m*x + t
```

Hier nutzen wir einen sogenannten "Docstring", abgegrenzt durch """. Dieser beschreibt die was unsere Funktion für uns erledigt, welche Parameter sie annimmt und welche Werte sie ausgibt.

Wenn ihr auf einen Funktionsnamen klickt, könnt ihr den Docstring anschauen indem ihr "Shift+Tab" drückt.

```python
lin_function(1,1,1)
```

## 2. Lineare Funktion plotten

```python
xs = np.linspace(-10,10,100)

m = 2
t = 5

plt.plot(xs, lin_function(xs, m, t))
plt.ylabel('y')
plt.xlabel('x')

plt.show()
```

```python
xs = np.linspace(-10,10,100)

ms = [2, 3, -1]
ts = [5, 2, 6]

for m,t in zip(ms, ts):
    plt.plot(xs, lin_function(xs, m, t), label=f'y = {m} x + {t}')
    
plt.ylabel('y')
plt.xlabel('x')
plt.legend()

plt.show()
```

## 3. Überschrift und Achsen

```python
plt.plot(xs, lin_function(xs, m, t))
plt.ylabel('y')
plt.xlabel('x')

plt.title('Lineare Funktion')

plt.show()
```

## 4. Potenzgesetz

```python
def power_law(x, a, b):
    """
    Funktion für Potenzgesetz, gibt y = a * x**b aus.
        
    Parameter:
        x: Wert aus dem Definitionsbereich
        a: Koeffizient
        b: Exponent
        
    Returns:
        y: Funktionswert am Punkt x
        
    """
    
    return a* x**b
```

```python
xs = np.linspace(0.1,100,1000)

a = 1
b = 2.3

plt.plot(xs, power_law(xs, a, b))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Potenzgesetz')
plt.show()
```

## Doppelt logarithmischer Plot

```python
plt.plot(xs, power_law(xs, a, b))

plt.yscale('log')
plt.xscale('log')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Potenzgesetz')
plt.show()
```

## 6. interaktiver Plot

```python
from ipywidgets import interact, FloatSlider

def plot_power_law(a, b):
    x = np.linspace(0.1, 10, 100)
    y = a * x**b
    
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'y = {a} * x**{b}')
    plt.xlabel('x')
    plt.ylabel('y')
    
    plt.yscale('log')
    plt.xscale('log')
    
    plt.title('Power Law: y = a * x**b')
    plt.legend()
    plt.grid(True)
    plt.show()

# Create sliders for 'a' and 'b'
a_slider = FloatSlider(value=1.0, min=0.1, max=10.0, step=0.1, description='a:')
b_slider = FloatSlider(value=1.0, min=-5.0, max=5.0, step=0.1, description='b:')

# Use interact to create the interactive plot
plot = interact(plot_power_law, a=a_slider, b=b_slider)
```

Interpretation:

$$ y = a \cdot x^b \Rightarrow \log(y) = \log(a) + b \cdot \log(x) $$

Wir können $a$ aus dem y-Achsenabschnitt ermitteln und $b$ aus der Steigung!
